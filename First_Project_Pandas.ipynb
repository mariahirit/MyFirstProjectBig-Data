8.Aumentar el numero "MPG_CITY" con 5

```python
import random
def Aumentar(MPG_City):
  calcular = MPG_City * random.randint(3,5)
  return calcular

df["calcular"]= df["MPG_City"].apply(Aumentar)

df.head(2)
```

9.Agrupar por Make, en funcion de modelo de coches

```python
df.groupby("Make").mean()
```

```python
group = df.groupby("Make").agg({
    "EngineSize" :"sum",
    "Cylinders"  : "mean",
    "Weight"     : "max"
  
})
group[group["Weight"] > 7000]
```
